```python
def per(matrica):
    n=len(matrica)
    m=len(matrica[0])
    matrvib=[0]*(n*m)
    x=int(0)
    for i in range(0, n):
        for j in range(0, m):
            matrvib[x]=matrica[i][j]
            x=x+1
    return matrvib

def vivod(matrica, matrvib):
    n=len(matrica)
    m=len(matrica[0])
    x=int(0)
    matrix=[[0] * m for i in range(n)]
    for i in range(0, n):
        for j in range(0, m):
            matrix[i][j]=matrvib[x]
            x=x+1
    return matrix 
# Sortirovka viborom
def viborom(matrica):
    matrvib=per(matrica)
    for i in range(0, len(matrvib)):
        for j in range(i, len(matrvib)):
            if matrvib[i]>matrvib[j]:
                matrvib[i], matrvib[j]=matrvib[j], matrvib[i]     
    return vivod(matrica, matrvib)
# Konec viborom
# Sortirovka vstavkoy
def vstavka(matrica):
    matrvib=per(matrica)
    for i in range(1, len(matrvib)):
        j=i-1
        while (j>=0) and (matrvib[j]>matrvib[j+1]):
            matrvib[j], matrvib[j+1]=matrvib[j+1], matrvib[j]
            j=j-1
    return vivod(matrica, matrvib)


# Konec vstavkoy
# Sortirovka obmenom
def obmen(matrica):
    matrvib=per(matrica)
    l=len(matrvib)
    while l>1:
        for i in range(l-1):
            if matrvib[i]>matrvib[i+1]:
                matrvib[i], matrvib[i+1]=matrvib[i+1], matrvib[i]
        l=l-1
        
    return vivod(matrica, matrvib)
# Konec obmenom

# Sortirovka Shella
def shella(matrica):
    matrvib=per(matrica)
    length=len(matrvib)
    l=length
    while l>0:
        for i in range(length-l):
            j=i
            while j>=0 and matrvib[j+l]<matrvib[j]:
                    matrvib[j], matrvib[j+l]=matrvib[j+l], matrvib[j]
                    j=j-l
        l=int(l/2)
    return vivod(matrica, matrvib)
# Konec Shella

# Turnirnaya sortirovka
def turnirnaya(matrica):
    matrvib=per(matrica)
    def _tyrsort(matrvib):
        tree = [None] * 2 * (len(matrvib) + len(matrvib) % 2)
        index = len(tree) - len(matrvib) - len(matrvib) % 2
        for i, v in enumerate(matrvib):
            tree[index + i] = (i, v)
        for j in range(len(matrvib)):
            n = len(matrvib)
            index = len(tree) - len(matrvib) - len(matrvib) % 2
            while index > -1:
                n = (n + 1) // 2
                for i in range(n):
                    i = max(index + i * 2, 1)
                    if tree[i] != None and tree[i + 1] != None:
                        if tree[i][1] < tree[i + 1][1]:
                            tree[i // 2] = tree[i]
                        else:
                            tree[i // 2] = tree[i + 1]
                    else:
                        tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
                index -= n

            index, x = tree[0]
            matrvib[j] = x
            tree[len(tree) - len(matrvib) - len(matrvib) % 2 + index] = None
    _tyrsort(matrvib)
    return vivod(matrica, matrvib) 
# Konec turnirnoy

# Bistraya sortirovka
def bs(matrica):
    matrvib=per(matrica)
    def bistraya(matrvib, first, last):
        i=first
        j=last
        x=matrvib[int((first+last)/2)]
        while i<=j:
            while matrvib[i]<x:
                i+=1
            while matrvib[j]>x:
                j-=1
            if i<=j:
                if matrvib[i]>matrvib[j]:
                    matrvib[i], matrvib[j]=matrvib[j], matrvib[i]
                i+=1
                j-=1
        if i<last:
            bistraya(matrvib, i, last)
        if first<j:
            bistraya(matrvib, first, j)
    bistraya(matrvib, 0, (n*m)-1)
    return vivod(matrica, matrvib)
# Konec Bisroy

# Piramidal'naya sortirovka
def heapSort(matrica):
    matrvib=per(matrica)
    def downHeap(matrvib, k, n):
        el = matrvib[k]
        while 2*k+1 < n:
            child = 2*k+1
            if child+1 < n and matrvib[child] < matrvib[child+1]:
                child += 1
            if el >= matrvib[child]:
                break
            matrvib[k] = matrvib[child]
            k = child
        matrvib[k] = el
  
    size = len(matrvib)
    for i in range(size//2-1,-1,-1):
        downHeap(matrvib, i, size)
    for i in range(size-1,0,-1):
        temp = matrvib[i]
        matrvib[i] = matrvib[0]
        matrvib[0] = temp
        downHeap(matrvib, 0, i)
    return vivod(matrica, matrvib)
# Konec piramidal'noy

# Zadanie №2         
import numpy
import time
def matrixgenerate():
    n=int(input())
    m=int(input())
    matrica = numpy.random.random_integers(-250, 1017, size=(n, m))
    print(*matrica)
    return matrica
        
# Vivod
matrica=matrixgenerate()

#Стандартная сортировка
timerStart = time.perf_counter()
matrvib=per(matrica)
matrvib=sorted(matrvib)
matrixstandart=vivod(matrica, matrvib)
timerEnd = time.perf_counter()
timer1 = timerEnd - timerStart

#Сортировка выбором
timerStart = time.perf_counter()
matrvibor=viborom(matrica)
timerEnd = time.perf_counter()
timer2 = timerEnd - timerStart

#Сортировка вставками
timerStart = time.perf_counter()
matrvstavka=vstavka(matrica)
timerEnd = time.perf_counter()
timer3 = timerEnd - timerStart

#Сортировка обменом
timerStart = time.perf_counter()
matrobmen=obmen(matrica)
timerEnd = time.perf_counter()
timer4 = timerEnd - timerStart

#Сортировка Шелла
timerStart = time.perf_counter()
matrshell=shella(matrica)
timerEnd = time.perf_counter()
timer5 = timerEnd - timerStart

#Турнирная сортировка
timerStart = time.perf_counter()
matrtur=turnirnaya(matrica)
timerEnd = time.perf_counter()
timer6 = timerEnd - timerStart

#Быстрая сортировка
timerStart = time.perf_counter()
matrbistraya=bs(matrica)
timerEnd = time.perf_counter()
timer7 = timerEnd - timerStart

#Пирамидальная сортировка
timerStart = time.perf_counter()
matrpir=heapSort(matrica)
timerEnd = time.perf_counter()
timer8 = timerEnd - timerStart




print("\nВремя работы стандартной сортировки", timer1,"\nRezultSort\n", *matrixstandart)
print("\nВремя работы сортировки выбором", timer2,"\nRezultVibor\n", *matrvibor)
print("\nВремя работы сортировки вставкой", timer3,"\nRezultVstavka\n", *matrvstavka)
print("\nВремя работы сортировки обменом", timer4,"\nRezultObmen\n", *matrobmen)
print("\nВремя работы сортировки Шелла", timer5,"\nRezultShella\n", *matrshell)
print("\nВремя работы турнирной сортировки", timer6,"\nRezultTurnirnaya\n", *matrtur)  
print("\nВремя работы быстрой сортировки", timer7,"\nRezultBistraya\n", *matrbistraya)
print("\nВремя работы пирамидальной сортировки", timer8,"\nRezultPiramidoy\n", *matrpir)
```
